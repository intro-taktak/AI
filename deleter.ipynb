```python
import os
import glob
import numpy as np
from PIL import Image

count = 0
os.chdir('C:\Projects\keras_talk\dataset\suit')
img_list = glob.glob('*.*')

for i in range(len(img_list)):
    im = Image.open(img_list[i])
    
    if(im.size[0] * 2 < im.size[1]) :
        im.close()
        os.remove(img_list[i])
        print('remove : ' + img_list[i])
        count = count + 1
        
print(str(count) + '개 항목 삭제 완료')
```
